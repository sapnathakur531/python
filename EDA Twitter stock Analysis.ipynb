# Z-Test or T-Test


```python
import statsmodels.api as sm
import scipy.stats as stats
```

```python
sample_high = data['High']
sample_low = data['Low']
sample_close = data['Close']

# Perform Z-test
z_statistic, p_value = sm.stats.ztest(sample_high, sample_low)

# Perform T-test
t_statistic, p_value = stats.ttest_ind(sample_high, sample_low)

alpha = 0.05

if p_value < alpha:
    print("Reject the null hypothesis")
else:
    print("Accept the null hypothesis")
```

# ANOVA

```python
#ANOVA
import pandas as pd
from scipy.stats import f_oneway

data['Year'] = data['Date'].dt.year

# Perform ANOVA
f_statistic, p_value = f_oneway(data['Open'], data['High'], data['Close'])

alpha = 0.05

if p_value < alpha:
    print("Reject the null hypothesis")
else:
    print("Accept the null hypothesis")

print("F-value:", f_statistic)
print("p-value:", p_value)


```

# CHI-SQUARE METHOD

```python

import pandas as pd
from scipy.stats import chi2_contingency

contingency_table = pd.crosstab(data['High'], data['Low'])

# Perform chi-square test
chi2_stat, p_val, dof, expected = chi2_contingency(contingency_table)

alpha = 0.05

if p_val < alpha:
    print("The variables are dependent")
else:
    print("The variables are independent")

print("Chi-Square Statistic:", chi2_stat)
print("p-value:", p_val)
print("Degrees of Freedom:", dof)
print("Expected Frequencies Table:\n", expected)
```

# z-score & outliers

```python
import pandas as pd
import numpy as np

column_to_analyze = 'High'

# Calculate mean and standard deviation
mean = data['High'].mean()
std_dev = data['High'].std()

# Calculate Z-scores for each data point
z_scores = (data['High'] - mean) / std_dev

# Define a Z-score threshold for outliers (e.g., Z > 3 or Z < -3)
z_score_threshold = 3

# Identify outliers based on the threshold
outliers = data[abs(z_scores) > z_score_threshold]

print("Outliers:")
print(outliers)
```

```python
# Create a function to update the plot based on the selected time period
def update_plot(column, period):
    plt.figure(figsize=(10, 6))
    data_period = data[column].resample(period).mean()  # Resample data based on the selected period
    plt.plot(data_period.index, data_period, label='column', color='blue')
    plt.xlabel('Date')
    plt.ylabel('column')
    plt.title('Twitter Stock {column} Over Time ({period} period)')
    plt.legend()
    plt.grid(True)
    plt.show()
    
# Create buttons for different columns and time periods

column_buttons = widgets.Dropdown(
    options=data.columns,
    description='Select Column:',
    disabled=False
)
period_buttons = widgets.RadioButtons(
    options=['D', 'W', 'M', 'Q', 'Y'],
    description='Time Period:',
    disabled=False
)
```

```python
plt.figure(figsize=(10,5))
plt.plot(df.index,df['Sale Price'],df['Mrp'], 'm-', linewidth=1.5, label='MRP')
plt.xlabel('sales and Mrp')
plt.ylabel('values')
plt.title('Sales & MRP')
plt.legend()
plt.show()
```
